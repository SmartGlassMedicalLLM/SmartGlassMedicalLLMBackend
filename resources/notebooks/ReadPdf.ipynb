```python
from PyPDF2 import PdfReader
from IPython.display import display, Markdown

def extract_text_pypdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() or "" # use or "" to handle potential None returns
    return text

# Example usage
pdf_text = extract_text_pypdf("/content/epi.pdf")
display(Markdown(pdf_text[0:800]))
```

```python
keyword = "interaction"

index = pdf_text.find(keyword)

bound_lower = index-100
bound_upper = index+100

pdf_text[bound_lower:bound_upper]
```
