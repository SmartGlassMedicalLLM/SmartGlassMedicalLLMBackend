```python
from PyPDF2 import PdfReader

def extract_text_pypdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() or "" # use or "" to handle potential None returns
    return text

# Example usage
pdf_text = extract_text_pypdf("/content/epidiolex.en.USPI.pdf")
print(pdf_text[0:100])
```

To search pdf for

```python
keyword = "interaction"

index = pdf_text.find(keyword)

bound_lower = index-100
bound_upper = index+100

pdf_text[bound_lower:bound_upper]
```
